```python
df_agrupado = pd.DataFrame(columns=['UF','2016','2017','2018','2019', '2020'])
for uf in df_2.ESTADO_UF.unique():
  row = pd.DataFrame([
        [uf, 
         df_2.QTD[(df_2.ANO==2016) & (df_2.ESTADO_UF == uf)].sum(),
         df_2.QTD[(df_2.ANO==2017) & (df_2.ESTADO_UF == uf)].sum(),
         df_2.QTD[(df_2.ANO==2018) & (df_2.ESTADO_UF == uf)].sum(),
         df_2.QTD[(df_2.ANO==2019) & (df_2.ESTADO_UF == uf)].sum(),
         df_2.QTD[(df_2.ANO==2020) & (df_2.ESTADO_UF == uf)].sum()
        ],
        ], columns=['UF','2016','2017','2018','2019', '2020'])
  df_agrupado = pd.concat([df_agrupado, row], axis=0,  ignore_index = True)

row = pd.DataFrame([
      ['Brasil', 
        df_2.QTD[(df_2.ANO==2016)].sum(),
        df_2.QTD[(df_2.ANO==2017)].sum(),
        df_2.QTD[(df_2.ANO==2018)].sum(),
        df_2.QTD[(df_2.ANO==2019)].sum(),
        df_2.QTD[(df_2.ANO==2020)].sum()
      ],
      ], columns=['UF','2016','2017','2018','2019', '2020'])
df_agrupado = pd.concat([df_agrupado, row], axis=0,  ignore_index = True)
  
```

```python
df_agrupado.head(28)
```

```python
from scipy.stats import chisquare

df_agrupado['Qui-Quadrado'] = 0
df_agrupado['p-value'] = 0
for idx in range(0, len(df_agrupado.UF)-1):
  
  A_2016 = df_agrupado['2016'].iloc[idx]
  A_2017 = df_agrupado['2017'].iloc[idx]
  A_2018 = df_agrupado['2018'].iloc[idx]
  A_2019 = df_agrupado['2019'].iloc[idx]
  A_2020 = df_agrupado['2020'].iloc[idx]

  CHI, P_VALUE = chisquare([A_2016, A_2017, A_2018, A_2019, A_2020])
  df_agrupado['Qui-Quadrado'].loc[idx] = CHI
  df_agrupado['p-value'].loc[idx] = P_VALUE

  #idx+=1
```

```python
def qui_quadrado(obs, exp):
  x2_calculado = 0
  for idx in range(0, len(obs)):
    o = obs[idx]
    e = exp[idx]
    x2_calculado += (((o-e)**2)/e)
  return x2_calculado
```
